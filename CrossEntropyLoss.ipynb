```python
import math

inputs = [(0.26, 1),
          (0.20, 0),
          (0.48, 1),
          (0.30, 0)]

def cross_entropy_loss(inputs):

    loss = 0
    n = len(inputs)

    for input in inputs:
        w_sum = input[0]
        y = input[1]

        loss += -(y * math.log10(w_sum) + (1 - y) * math.log10(1 - w_sum))
        print(-(y * math.log10(w_sum) + (1 - y) * math.log10(1 - w_sum)))

    return loss / n

error_term = cross_entropy_loss(inputs)
print(f"Error term: {error_term}")
```
